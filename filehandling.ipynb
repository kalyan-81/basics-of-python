# Opening a file
we should use **open()** function to open a file.
*syntax


```
file=open('filename','openmode','buffering')
```
filename is the name of the file to be open.\
open mode represents the purpose of the opening file.


```
open mode                    description
    w            To write the data into file, if any data already present in the file. it would be deleted and the present data will be stored.
    r           To read the data from the file. The file pointer is positioned at the begining of the file.
    a           To append data to the file. Appending means adding at the end of the existing data. The file pointer is palced at the end of the file. If the file doesn't exist, it will create a new file for writing data.
    w+         To write and read data of a file. The previous data in the file will be deleted.
    r+         To read and write the data into file. The previous data in the file will not be deleted. The file pointer is placed at the beginning of the file.
    a+         To append and read data of a file. The file pointer will be at the end of the file if the file exists. If the file doesn't exist, it will create a new file for reading and writing.
    x         To open the file in exclusive creation mode. The file creation fails if the file already exists.
```







A **buffer** represents a temporary block of memory. buffering is an optional integer used to set the size of the buffer for the file. In the binary, we can pass 0 as buffering integer to inform not to use any buffering.

In text mode, we can use 1 for buffering to retrieve the data from the file one line at a time.

We can use any integer for buffering.
If we do not mention buffersize, the default buffersize used is 4096 or 8192 bytes.

# Pickle in Python.
upto now we stored text data into file.\
**what happens when we want to store some structured data in the file?.**\
for example, we want to store some employee data like id number(int type), name(str type), and salary(float type) in a file.\
this data is well structured and got different types.\
so to store such data, we need to create a class Employee with instance variables id, name, and sal as shown here.


````python
class Emp:
  def __init__(self,id,name,sal):
    self.id=id
    self.name=name
    self.sal=sal
  def display(self):
    print('{:5d} {:20s}{:10.2f}'.format(self.id,self.name,self.sal))
````

Now we can create an object to this class and store actual data into that object. Later, this object should be stored into a binary file in the form of bytes. This is called pickle or serialization.

**pickle** is a process of converting a class object into a byte stream so that it can be stored into a file. This is also called object serialization.

**pickling is done using dump() method of 'pickle' module as:**

* syntax\
pickle.dump(object, file)

**unpickling is done using load() method of 'pickle' module as**:
* syntax\
object=pickle.load(file)
unpickling is also called de-serialization.

````python
e=Emp(100,"kalyan",50000)
````

````python
import pickle
f=open('emp.dat','wb') #emp.dat is file which stores binary data
pickle.dump(e,f)# the object e is converted to binary stream and stores in the file 'f'
f.close()
````

````python
#unpickling
f=open('emp.dat','rb')
obj=pickle.load(f)# unpickling using load() method
obj.display() #calling function
f.close()
````

````python
from posixpath import normpath
#write a python program to pickle Emp class objects
import pickle
f=open('emp.dat','wb')
n=int(input('number of employees:'))
for i in range(n):
  id=int(input('enter id number:'))
  name=input('enter your name:')
  sal=int(input('enter salary:'))
# create object for above emp class
  e=Emp(id,name,sal)
# store the object into file f
  pickle.dump(e,f)
f.close()
````

````python
# unpickling for the above one
import pickle
f=open('emp.dat','rb')
print('employees details: ')
while True:
  try:
    #read object from file f
    obj=pickle.load(f)
    #display the contents of employee obj
    obj.display()
  except EOFError:
    print('end of the file reached')
    break
f.close()
````

# Random accessing of binary files
let's us take binary file having 1000 bytes of data. if we want to access the last 10 bytes of data, it is not needed to search the file byte by byte from the beginning. it is possible to directly go to 991 position uisng the seek() method as.

* f.seek(900)
then the last 10 bytes using the read() method as:
* f.read(10)

In this way we can directly going to any byte in the binary file is called random accessing.

* binary files accepts data in the form of bytes or in binary format.

````python
# example 
str1='master python'
with open('data.bin','wb') as f:
  f.write(str1) #store str1 into binary file
  f.write('hello') # store hello
# it will produce an error since string is not binary. we have to convert to binary file
````

````python
# create a binary file and store a few records.
#create cities.bin file with cities names
#take the record size as 20 bytes
reclen=20
#open the file in wb mode as binary file
with open('cities.bin','wb') as f:
  #to write data into entries
  n=int(input('how many entries?'))
  for i in range(n):
    city=input('enter city name:')
    #find length of city
    ln=len(city)
    #increase the city name to 20 characters
    #by adding remaining spaces
    city=city +(reclen-ln)*' '
    #convert the city name into the byte string
    city=city.encode()
    #write the city name into the file
    f.write(city)
````

````python
# program to randomly access the record from the cities.bin file
with open('cities.bin','rb') as f:
  n=int(input('enter the record numbr:'))
  # to move the file pointer to the end of the (n-1)th record
  f.seek(reclen*(n-1))
  #get the nth record with 20 chars
  str1=f.read(reclen)
  print(str1)
  #convert the byte string into ordinary string
  print(str1.decode())
````

````python
#program to update or modify a record in a binary file
import os
#take the record length as 20 characters
reclen=20
#find the size of the file
size=os.path.getsize('cities.bin')
print('size of the file ={}'.format(size))
#find number of records in the file
n=int(size/reclen)
print('No. of records={}'.format(n))
# open the file in binary mode for reading
with open('cities.bin','r+b') as f:
  name=input('enter the city name:')
  #convert the name into binary string
  name=name.encode()
  newname=input('enter new name:')
  #find the length of new name
  ln=len(newname)
  # add spaces to make its length to 20 characters
  newname=newname+(20-ln)*' '
  #convert the newname into binary string
  newname=newname.encode()
  # position represents teh position o f file pointer
  position=0
  #found becomes True if city is found in the file
  found=False

  #repeat for n records in the file
  for i in range(n):
    #place the file pointer at position
    f.seek(position)
    #read 20 charactes
    str1=f.read(20)
    if name in str1:
      print('updated record no:',(i+1))
      found=True
      #go back to 20 bytes from current positions
      f.seek(-20,1)
      #update the record
      f.write(newname)
    #go to the next record
    position+=reclen
  if not found:
    print('city not found')
````

we cannot delete a record from the binary file. To delete a record follow the below rules\
* copy the all record except the record to be delete into another file
* remove the original file using os.remove() method
* rename the newfile to original file using os.rename()

````python
# A program to delete a specific record from the binary file
import os
# take record length as 20 characters
reclen=20
#find the Size of the file
size=os.path.getsize('cities.bin')
#find number of records in the file
n=int(size/reclen)
#open the cities.bin for reading
f1=open('cities.bin','rb')
#open file2.bin for writing
f2=open('file2.bin','wb')
# accept city  name from keyboard
city=input('enter city name to delete')
# add spaces so that it will have 20 characters length
ln=len(city)
city=city + (reclen-ln)*' '
#convert city name to binary string
city=city.encode()
#repeat for all the n records
for i in range(n):
  #read one record from f1 file
  str1=f1.read(reclen)
  #if it is not the city name, store into f2 file
  if(str1!=city):
    f2.write(str1)
print('record deleted')
#close the files
f1.close()
f2.close()
````
