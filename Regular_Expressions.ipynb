#Methods belong to **re** module that are used in the regular expressions:
**match()** method: searches in the beginning(only starting) of the string and if the matching string is found, it returns  the resultant string, otherwise it returns None. we can access the string from the returned object using **group()** method.

**search()** method: searches the string from beginning to end of the string and returns an object that contains the resultant string, otherwise it returns None. we can access the string from the returned object using **group()** method.

**findall()** method: searches the string from beginning till the end and return all occurences of the matching string in the form of a list object. if the  matching strings are not found it returns an empty list.

**split()**method: splits the string according to the regular expression and the resultant pieces are returned as a list. If there are no string pieces it reurns an empty list.

**sub()** method: sustitutes(or replaces) new strings in the place of existing strings. after substitution , the main string is returned by this method.

````python
#exmaple for search()
import re
str1='man sun mop run'
result=re.search(r'm\w\w',str1) # this method returns None if string not matches with expression
if result:
  print(result.group()) #if result is not None
````

````python
#exmaple for search()
import re
str1=' sun man mop run'
result=re.search(r'm\w\w',str1) # this method returns None if string not matches with expression
if result:
  print(result.group()) #if result is not None
````

````python
#exmaple for match()
import re
str1='man sun mop run'
result=re.match(r'm\w\w',str1) # this method returns None if string not matches with expression
if result:
  print(result.group()) #if result is not None
````

````python
#exmaple for match(). this will search only starting. it will not search upto end.
import re
str1='  man sun mop run'
result=re.match(r'm\w\w',str1) # this method returns None if string not matches with expression
if result:
  print(result.group()) #if result is not None
else:
  print('not found')
````

````python
#example for findall(). this will return the matched string in a list

import re
str1='  man sun mop run'
result=re.findall(r'm\w\w',str1) # this returns a list of matched strings
print(result)
````

````python
# example for split() method
import re
str1='this; is the : master python channel'
result=re.split(r'\W+',str1) # W is uppercase. + represents to match 1 or more occurrences indicated by W.
print(result)
````

# sequence characters in Regular expressions
sequence characters match only one character in the string.


```
character                           description

\d                           represents any digit(0 to 9).
\D                           represents any non-digit.
\s                           represents white space. ex: \t,\n, \r, \f, \v.
\S                           represents non-white space character.
\w                           represents any alphanumeric(A to Z, a to z, 0 to 9)
\W                           represents non-alphanumeric.
\b                           represents a space around words.
\A                           matches only at start of the string
\Z                           matches only at the end of the string
```
each of these sequence characters represents a single character matched in the string.
suppose '\w' represents any one alphanumeric character.
suppose if write [\w]*, here *  represents 0 or more repetitions. hence it will represents 0 or more alphanumeric characters.



````python
# regular expression to retrieve all words starting with 'a' given string.
import re
str1='an apple a day keeps the doctor away'
result=re.findall(r'a[\w]*',str1)
print(result)
````

please observe the output. It contains 'ay' which is not a word. This 'ay' is part of the word 'away'. So, it is displaying both 'ay' and 'away' as they are starting with 'a'. we do not want like this. we want only the words starting with 'a'. Since a word will have a space in the beginning or ending, we can use '\b' before and after the words in the regular expression. So regular expression will become.


````python
#import re
str1='an apple a day keeps the doctor away'
result=re.findall(r'\ba[\w]*\b',str1)
print(result)
````

````python
# regular expression to retrieve all the words starting with a numeric digit
import re
str1='The meeting will be conducted on 1st and 21st of every month'
result=re.findall(r'\d[\w]*',str1) # \d represents any digit
print(result)
````

````python
# regular expression to retrieve all words having 5 characters length
import re
str1='one two three four five six seven 8 9 10'
result=re.findall(r'\b\w{5}\b',str1)
print(result)
````

````python
# # regular expression to retrieve all words having 5 characters length using search method
import re
str1='one two three four five six seven 8 9 10'
result=re.search(r'\b\w{5}\b',str1)
print(result.group())
````

````python
# regular expression to retrieve all the words that are having the length of at least 4 characters
import re
str1='one two three four five six seven 8 9 10'
result=re.findall(r'\b\w{4,}\b',str1)
print(result)
````

````python
# regular expression to retrieve all words with 3 or 4 or 5 characters length
import re
str1='one two three four five six seven 8 9 10'
result=re.findall(r'\b\w{3,5}\b',str1) # this expression will three 3 to 5 characters
print(result)
````

````python
# regular expression to retrieve only single digits from the given string
import re
str1='one two three four five six seven 8 9 10'
result=re.findall(r'\b\d\b',str1)
print(result)
````

````python
# regular expression to retrieve the last word of a string, if it starts with 't'.
import re
str1='one two three four five six seven 8 9 10 thirteen'
result=re.findall(r't[\w]*\Z',str1) 
print(result)
````

# Quantifiers in Regular Expressions
some characters represent more than one character to be matched in the string. Such characters are called 'Quantifiers'.


```
character                     description
  +                      1 or more repetitions of the preceding regexp.
  *                      0 or more repetitions of the preceding regexp.
  ?                      0 or 1 repetitions of the preceding regexp.
 {m}                    exactly m occurrences.
 {m,n}                  from m to n occurrences, defualt m is 0 and n is infinity
```



````python
#regular expression to retrieve the phone number of a person.
import re
str1='Nageswara Rao: 9706612345'
res=re.search(r'\d+',str1)
print(res.group())
````

````python
#regular expression to retrieve the non digit characters of a person.
import re
str1='Nageswara Rao: 9706612345'
res=re.search(r'\D+',str1)
print(res.group())
````

````python
# regular expression to find all words starting with 'an' or 'ak', then we can use
import re
str1='anil akhil anant arun arati arundhati abhijit ankur'
res=re.findall(r'a[nk][\w]*',str1) # [nk] represents either n or k
print(res)
````

````python
# regular expression to retrieve the date of births from a string
import re
str1='kalyan 24 26-08-1997, rohit 21 22-10-1990, sita 22 15-09-2000'
res=re.findall(r'\d{2}-\d{2}-\d{4}',str1)
print(res)
````

# special characters in Regular expressions


```
character                  description
  \                    escape special character nature
  .                    matches any character except new line
  ^                    matches beginning of a string, useful to check substring or not
  $                    matches ending of a string
  [...]                denotes a set of possible characters. ex: [6c-f] matches any characters '6','c','d','e','f'.
  [^...]               matches every character except the ones inside brackets. ex: [^6c-f] matches any characters except '6','c','d','e','f'.
  (...)                matches the regular expression inside parentheses and the result can be captured.
  R | S               matches either regex R or regex 
  
```



````python
#regex to check whether a given string is starting with 'He' or not
import re
str1='hello world'
res=re.search(r'^he',str1)
if res:
  print('string starts with he')
else:
  print('string does not start with he')
````

````python
# regex to search for a word at the ending of a string
import re
str1='Hello world'
res=re.search(r'world$',str1)
if res:
  print('string ends with "world"')
else:
  print('string does not end with "world"')
````

````python
#program to search at the ending of the string by ignoring the case
import re
st1='HELLO WORLD'
res=re.search(r'world$',st1,re.IGNORECASE)
if res:
  print('string ends with "world"')
else:
  print('string does not end with "world"')
````

The square bracket [ ] in regular expressions represent a set of characters. For example if we write [ABC], it represents any one character A or B or C. A hyphen (-) represents range of characters .ex: [a - z] represents any single lower case character from the range of a to z.

````python
#regular expression to retrieve the marks and names from a given string
import re
str1='Rahul got 75 marks,Vijay got 55 marks, whereas Subbu got 98 marks'
#extract only marks having 2 digits
marks=re.findall('\d{2}',str1)
print(marks)
#extracting the names starting with capital letters
names=re.findall(r'[A-Z][a-z]*',str1)
print(names)
````

The pipe symbol (|) represents 'or'. ex: if we write 'am | pm' , it finds the strings which are either 'am' or 'pm'

````python
import re
str1='The  meeting may be at 8am or 9am or 4pm or 5pm'
res=re.findall(r'\dam|\dpm',str1)
print(res)
````

# Using Regular expression on Files




````python
#let'us create a text file
f=open('mails.txt','a')
line1='hi my mail id is kalyan@gmail.com\n'
line2='hi raju\'s mail id is raju12@yahoo.com\n'
line3='hi yours mail id is yoursname34@gmail.com\n'
f.writelines(line1)
f.writelines(line2)
f.writelines(line3)
f.close()
````

````python
# regular expression to read emails ids from a text file
import re
#open file in reading mode
f=open('mails.txt','r')
# search in each line of the file
for line in f:
  res=re.findall(r'\S+@\S+',line)
# display if there are some elements in result
  if len(res)>0:
      print(res)
#close the file

f.close()
````
